```python
import os
import pandas as pd
import json
import warnings
warnings.filterwarnings("ignore")
import nltk   
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
def getHtmlText(tUrl):
    url1 = tUrl
    html = urlopen(url1).read()
    soup = BeautifulSoup(html, features="html.parser")
    # kill all script and style elements
    for script in soup(["script", "style"]):
        script.extract()    # rip it out
    # get text
    text = soup.get_text()
    # break into lines and remove leading and trailing space on each
    lines = (line.strip() for line in text.splitlines())
    # break multi-headlines into a line each
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    # drop blank lines
    text1 = '\n'.join(chunk for chunk in chunks if chunk)
    return text1
```

```python
# PATH TO DOWNLOADED DATA
directory = '/notebooks/data/Euro-stat-sample-data/'

count = 0
for filename in os.listdir(directory):
    final_df = pd.DataFrame(columns=['TITLE1:', 'TITLE2:', 'TITLE3:'])
    f = os.path.join(directory, filename)
    fchunk = filename.split("_")
    ref_text = ""
    linkToReferenceMD = "https://ec.europa.eu/eurostat/cache/metadata/en/"+fchunk[0]+"_esms.htm"
    print(linkToReferenceMD)
    try:
        ref_text  = getHtmlText(linkToReferenceMD)
        print(str(count))
        print(fchunk[0])
    except:
        ref_text = ""
        
    filename = filename.replace(".tsv.gz", "")
    new_df = pd.read_csv(f, sep='\t')
    cols_list = new_df.columns.tolist()
    
    if len(cols_list) > 0: 
        listToStr = ','.join([str(elem) for elem in cols_list])
        final_df = final_df.append({"TITLE1:":" The description of the data in the table "+ filename, "TITLE2:": " The description of columns data " + listToStr, "TITLE3:": " The general description of the table data "+ref_text}, ignore_index = True)
        df2 = final_df.to_json(orient = 'records')
        with open(os.path.join("/notebooks/data/sample-json-data/", filename+".json"), "w") as outfile:
             outfile.write(df2)
    count = count + 1   
```
